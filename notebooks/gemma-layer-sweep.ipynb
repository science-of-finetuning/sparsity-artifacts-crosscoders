```python
import pandas as pd
import plotly.express as px
import json
from pathlib import Path

file = "/dlabscratch1/jminder/repositories/representation-structure-comparison/feature_ablation_results/*sweep_switch_ablation_None_to_base_switch.json"
with open(file, "r") as f:
    data = json.load(f)

data
```

```python
data["1"][1]["steered"][1]["content"]
```

```python
# Extract the relevant content from the data
question_idx = 1
steered_contents = [entry[question_idx]['steered'][1]['content'] for entry in data.values()]

# Create a DataFrame to hold the binary values
df = pd.DataFrame({
    'gemma': [1 if 'gemma' in content.lower() else 0 for content in steered_contents],
    'openai': [1 if 'openai' in content.lower() else 0 for content in steered_contents],
    'gpt': [1 if 'gpt' in content.lower() else 0 for content in steered_contents],
    'deepmind': [1 if 'deepmind' in content.lower() else 0 for content in steered_contents],
    "google": [1 if "google" in content.lower() else 0 for content in steered_contents]
})

# Plot the lines
fig = px.line(df, title=f"Occurrences of Keywords in Responses for '{data['1'][question_idx]['steered'][0]['content']}'")
fig.update_layout(yaxis_title='Appearances', xaxis_title='Switch Layer')
fig.write_image("gemma_layer_sweep.png", scale=3)
fig.show()
```
