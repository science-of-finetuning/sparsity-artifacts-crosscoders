## This notebook formats the LMSYS chat dataset into text using the gemma chat template

```python
from datasets import load_dataset
from transformers import AutoTokenizer
from tqdm.notebook import tqdm
from pathlib import Path
from collections import defaultdict

DATA_PATH = Path("../datasets/")
source_dataset = "science-of-finetuning/lmsys-chat-1m-gemma-formatted"
target_dataset = "science-of-finetuning/lmsys-chat-1m-gemma-formatted"
```

## Make dataset

```python
tokenizer = AutoTokenizer.from_pretrained("google/gemma-2-2b-it")
ds = load_dataset(source_dataset)
```

```python
formated_conversations = defaultdict(list)
for split in ds.keys():
    for conversation in tqdm(ds[split]["conversation"]):
        formated_conversations[split].append(
            tokenizer.apply_chat_template(conversation, tokenize=False)[5:]
        )
```

```python
def apply_connor_template(conv):
    return (
        "\n".join(
            [
                ("Assistant: " if msg["role"] == "assistant" else "User: ")
                + msg["content"]
                for msg in conv
            ]
        )
        + "\n"
    )


formated_conversations_connor = {
    split: [apply_connor_template(conv) for conv in tqdm(ds[split]["conversation"])]
    for split in ds.keys()
}
```

```python
print(formated_conversations_connor[30])
print("=" * 100)
print(formated_conversations[30])
```

```python
# create a new dataset
ds_formatted = ds
for split in ds.keys():
    ds_formatted[split] = (
        ds_formatted[split]
        # .add_column("text", formated_conversations[split])
        .add_column("text_base_format", formated_conversations_connor[split])
    )
# Save the new dataset to disk
# ds_formatted.save_to_disk(DATA_PATH / "lmsys-chat-1m-formatted")
# ds_formatted.push_to_hub(target_dataset)
```
